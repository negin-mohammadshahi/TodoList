```python
(X_train2, y_train2), (X_test2, y_test2) = keras.datasets.imdb.load_data(num_words=10000)
X_train2 = keras.preprocessing.sequence.pad_sequences(X_train2, maxlen=200)
X_test2 = keras.preprocessing.sequence.pad_sequences(X_test2, maxlen=200)

model2 = keras.Sequential([
    layers.Embedding(10000, 32),
    layers.GlobalAveragePooling1D(),
    layers.Dense(16, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model2.fit(X_train2, y_train2, epochs=10, batch_size=512, validation_split=0.2, verbose=0)

loss, accuracy = model2.evaluate(X_test2, y_test2, verbose=0)
print(f"Test accuracy: {accuracy * 100:.2f}%")

predictions = model2.predict(X_test2[:5], verbose=0)
for i in range(5):
    score = predictions[i][0]
    actual = "Positive" if y_test2[i] == 1 else "Negative"
    predicted = "Positive 😊" if score > 0.5 else "Negative 😞"
    print(f"Review #{i+1} | Actual: {actual} | Predicted: {predicted} ({score:.2%})")
```
